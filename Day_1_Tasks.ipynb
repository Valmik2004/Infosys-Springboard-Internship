```python
# Task 1:
# Load the dataset and print:
# ● Total number of records
# ● Number of missing values per column
# ● Count of real vs fake jobs
# ● Display 3 examples of fake job .

#  Task 1
print("Load The Dataset")
import pandas as pd

# Replace the file name below with your actual uploaded file name
df = pd.read_csv("fake_job_postings.csv")

# Total number of records
print("Total records:", len(df))

# Missing values per column
print("\nMissing values per column:\n", df.isnull().sum())

#Count of real vs fake jobs
print("\nCount of real vs fake jobs:\n", df['fraudulent'].value_counts())

# Display 3 examples of fake job descriptions
print("\nThree examples of fake job descriptions:\n")
fake_jobs = df[df['fraudulent'] == 1]
for i, desc in enumerate(fake_jobs['description'].head(3), start=1):
    print(f"Fake Job {i}:\n{desc}\n{'-'*60}")

```
