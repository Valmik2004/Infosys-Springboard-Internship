```python
# Task 1:
# Write your own cleaning function that:
# ‚óè Removes HTML, numbers, and punctuation
# ‚óè Converts text to lowercase
# ‚óè Removes stopwords
# ‚óè Returns a cleaned version of the company_profile column
#-------------------------------------------------------------------------------
# Task 1: Custom Cleaning Function.
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Load dataset
df = pd.read_csv("fake_job_postings.csv")

# Define your own cleaning function
def custom_clean_text(text):
    if pd.isnull(text):
        return ""
    # 1. Lowercase
    text = text.lower()
    # 2. Remove HTML tags
    text = re.sub(r'<.*?>', ' ', text)
    # 3. Remove numbers and punctuation
    text = re.sub(r'[%s\d]' % re.escape(string.punctuation), ' ', text)
    # 4. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    # 5. Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in text.split() if word not in stop_words]
    return " ".join(words)

# Apply cleaning to 'company_profile'
df['clean_company_profile'] = df['company_profile'].apply(custom_clean_text)

# Display before and after cleaning for example
print("Original Company Profile:\n", df['company_profile'].iloc[1])
print("\nCleaned Company Profile:\n", df['clean_company_profile'].iloc[1])
```

```python
# Task 2: Word Count Comparison (Before vs After Cleaning).
# ‚óè Count the average number of words before and after cleaning for job descriptions.
#--------------------------------------------------------------------------------
import pandas as pd

#  Step 1: Define a helper function to count words
def word_count(text):
    """Returns the number of words in a text string."""
    if pd.isnull(text):
        return 0
    return len(text.split())

# Step 2: Safety check ‚Äî confirm the cleaned column exists
if 'clean_company_profile' not in df.columns:
    raise ValueError("‚ùå The column 'clean_company_profile' does not exist. "
                     "Please run Task 1 first to create it.")

# üßæ Step 3: Calculate word counts before and after cleaning
df['word_count_before'] = df['company_profile'].apply(word_count)
df['word_count_after']  = df['clean_company_profile'].apply(word_count)

# üìä Step 4: Compute average word counts
avg_before = df['word_count_before'].mean()
avg_after  = df['word_count_after'].mean()

# üìà Step 5: Display results clearly
print("=== Average Word Count Comparison ===")
print(f"Before Cleaning : {avg_before:.2f} words")
print(f"After Cleaning  : {avg_after:.2f} words")

# üîç Step 5: Show sample rows for quick inspection
print("\n=== Sample Comparison ===")
display(df[['company_profile', 'clean_company_profile',
            'word_count_before', 'word_count_after']].head(5))
```
